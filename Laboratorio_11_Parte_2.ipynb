1. Comparación de calificaciones de estudiantes en diferentes asignaturas
Se quiere comparar las calificaciones promedio de los estudiantes en distintas materias para detectar diferencias de rendimiento.

•	asignaturas = Matemáticas,Ciencias,Historia,Inglés,Arte

•	calificaciones = 85, 90, 78, 88, 92

```python
import matplotlib.pyplot as plt

# Datos
asignaturas = ['Matemáticas', 'Ciencias', 'Historia', 'Inglés', 'Arte']
calificaciones = [85, 90, 78, 88, 92]

# Crear el gráfico
plt.figure(figsize=(8, 5))
plt.bar(asignaturas, calificaciones, color=['blue', 'green', 'orange', 'purple', 'red'])

# Personalización del gráfico
plt.title('Comparación de calificaciones por asignatura', fontsize=14)
plt.xlabel('Asignaturas', fontsize=12)
plt.ylabel('Calificaciones', fontsize=12)
plt.ylim(70, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Linea promedio
promedio = sum(calificaciones) / len(calificaciones)
plt.axhline(y=promedio, color='black', linestyle='--', label=f'Promedio: {promedio:.2f}')
plt.legend()

# Mostrar el gráfico
plt.show()
```

La asignatura 'Historia' se encuentra muy por debajo del promedio de las calificaciones (86.60) con un 78. Se debe emplear formas de aumentar el desempeño en esta asignatura.

2. Comparación del tiempo de carga de diferentes páginas web
 Se quiere visualizar el tiempo promedio de carga de diferentes sitios web para identificar cuál es más rápido o lento.

•	sitios = Sitio A, Sitio B,Sitio C,Sitio D,Sitio E

•	tiempos = 1.2, 2.5, 0.9, 3.0, 1.8


```python
import matplotlib.pyplot as plt

# Datos
sitios = ['Sitio A', 'Sitio B', 'Sitio C', 'Sitio D', 'Sitio E']
tiempos = [1.2, 2.5, 0.9, 3.0, 1.8]

# Crear el gráfico
plt.figure(figsize=(8, 5))
plt.bar(sitios, tiempos, color=['cyan', 'orange', 'green', 'red', 'purple'])

# Personalización del gráfico
plt.title('Comparación del tiempo de carga promedio por sitio web', fontsize=14)
plt.xlabel('Sitios Web', fontsize=12)
plt.ylabel('Tiempo de Carga (segundos)', fontsize=12)
plt.ylim(0, 3.5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Mostrar el gráfico
plt.show()
```

Existe una variación significativa entre el sitio más rápido (Sitio C) y el más lento (Sitio D), de más del triple (3.33).

3. Relación entre el número de horas de estudio y el rendimiento académico
Se quiere ver si existe una correlación entre las horas de estudio y el rendimiento en los exámenes.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Datos simulados
horas_estudio = [2, 4, 6, 8, 10, 12, 14]
rendimiento = [50, 60, 70, 80, 85, 90, 95]

# Crear el gráfico de dispersión
plt.figure(figsize=(8, 5))
sns.scatterplot(x=horas_estudio, y=rendimiento, color='blue', s=100)

# Ajustar una línea de tendencia
sns.regplot(x=horas_estudio, y=rendimiento, scatter=False, color='red', line_kws={"lw": 2, "linestyle": "--"})

# Personalización del gráfico
plt.title('Relación entre horas de estudio y rendimiento académico', fontsize=14)
plt.xlabel('Horas de Estudio', fontsize=12)
plt.ylabel('Rendimiento en el Examen (%)', fontsize=12)
plt.grid(alpha=0.5)

# Mostrar el gráfico
plt.show()

# Calcular correlación
correlacion = np.corrcoef(horas_estudio, rendimiento)[0, 1]
print(f"Coeficiente de correlación: {correlacion:.2f}")
```

Existe una correlación positiva clara entre las horas de estudio y el rendimiento en los exámenes. A medida que aumentan las horas de estudio, el rendimiento también mejora

4. Distribución de salarios en tres departamentos distintos
Se tiene un conjunto de datos de salarios en diferentes departamentos y quieres visualizar la variabilidad y mediana de los salarios en cada uno.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Datos simulados
departamentos = ['Ventas', 'Marketing', 'TI']
salarios = [
    [4000, 4500, 5000, 5500, 6000],  # Ventas
    [3800, 4200, 4800, 5200, 5900],  # Marketing
    [4500, 4700, 5100, 5800, 6100]   # TI
]

# Crear un DataFrame para Seaborn
import pandas as pd
data = pd.DataFrame({
    'Departamento': ['Ventas'] * len(salarios[0]) +
                    ['Marketing'] * len(salarios[1]) +
                    ['TI'] * len(salarios[2]),
    'Salario': salarios[0] + salarios[1] + salarios[2]
})

# Crear el gráfico de caja (boxplot)
plt.figure(figsize=(8, 5))
sns.boxplot(x='Departamento', y='Salario', data=data, hue = 'Departamento', palette='Set2', legend=False)

# Personalización del gráfico
plt.title('Distribución de salarios por departamento', fontsize=14)
plt.xlabel('Departamento', fontsize=12)
plt.ylabel('Salario', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.show()
```

El boxplot muestra la mediana (línea dentro de cada caja), los rangos intercuartiles (los límites de la caja), y posibles valores atípicos (puntos fuera de los bigotes).
El análisis revela que Ventas tiene una mayor variabilidad, mientras que TI ofrece salarios más consistentes. Esto puede ser útil para comprender las políticas de compensación y planificar ajustes si es necesario.
