Se tiene un conjunto de datos sobre empleados que incluye su nombre, edad, salario y fecha de ingreso. Se requiere implementar algunos pasos de aseguramiento de la calidad de los datos, como la verificación de valores faltantes y la validación de formatos de datos.
1.	Identificar y contar los valores faltantes en cada columna.
2.	Verificar el formato de la columna 'Salario' para asegurar que todos los valores sean numéricos.
3.	Validar y estandarizar el formato de la columna 'Fecha Ingreso' a YYYY-MM-DD.

```python
import pandas as pd
import numpy as np

data = {
    'Nombre': ['Juan' , 'Ana' , 'Pedro' , 'María' , 'Luis' , 'Ana'],
    'Edad': [28, 34, None, 45, 38, 34],
    'Salario': ['3000', '4000', 'cinco mil', '4500', '4000', '5000' ],
    'FechaIngreso': ['2021-01-15' , '2020/03/12' , '2022-07-01' , '2021/12/01' , '2021-05-20', '2020-03-12']
}

df = pd.DataFrame(data)
print("Información completa:")
print(df)

# Retirar registros repetidos y arreglar índices
df = df.drop_duplicates('Nombre', keep='last')
df = df.reset_index(drop=True)
print("\nTabla sin nombres repetidos e índices corregidos:")
print(df)

# 1.	Identificar y contar los valores faltantes en cada columna.
valores_faltantes = df.isnull().sum()
print("\nValores faltantes por columna:")
print(valores_faltantes)

# Reemplazar 'NaN' en 'Edad' por la media de edades
df['Edad'] = df['Edad'].fillna(df['Edad'].mean())
print("\nTabla con valores de edad inválidos reemplazados por la media:")
print(df)

# 2.	Verificar el formato de la columna 'Salario' para asegurar que todos los valores sean numéricos.
salarios_incorrectos = df[~df['Salario'].str.replace('[^0-9]', '', regex=True).fillna('').str.isdigit()]
print("\nSalarios con formato incorrecto:")
print(salarios_incorrectos)

# Retirar los registros 'salarios incorrectos' del df y arreglar índices
df = df.drop(index = salarios_incorrectos.index)
df = df.reset_index(drop=True)
print("\nTabla sin salarios incorrectos e índices corregidos:")
print(df)

# Convertir 'Salario' en columna numérica
df['Salario'] = df['Salario'].astype(np.float64) # Puede usarse también .astype(int)
print('\nTabla con salarios numéricos:')
print(df)

# 3.	Validar y estandarizar el formato de la columna 'Fecha Ingreso' a YYYY-MM-DD.
df['FechaIngreso'] = df['FechaIngreso'].str.replace('/', '-')
df['FechaIngreso'] = pd.to_datetime(df['FechaIngreso'], errors='coerce', format='%Y-%m-%d')
print("\nDatos después de estandarizar 'FechaIngreso':")
print(df[['Nombre', 'FechaIngreso']])
```

Se tiene un conjunto de datos sobre empleados que incluye su nombre, edad, salario, puesto de trabajo y fecha de ingreso. Se requiere implementar aseguramiento de la calidad de los datos, verificando la completitud de los datos, la validez de los salarios y la consistencia de las fechas de ingreso.
1.	Identificar y contar los valores faltantes en cada columna.
2.	Verificar el formato de la columna "Salario" para asegurarse de que todos los valores sean numéricos.
3.	Validar y estandarizar el formato de la columna "FechaIngreso" a YYYY-MM-DD.
4.	Verificar que no haya valores faltantes en la columna "Cargo".


```python
import pandas as pd
import numpy as np

data = {
    'Nombre': ['Juan' , 'Ana' , 'Pedro' , 'María' , 'Luis' , 'Carla '],
    'Edad': [28, None, 40, 45, 38, 34],
    'Salario': ['3000', '4000', '4500', 'cuatro mil', '5000' , '4000'],
    'Cargo' : ['Analista' , 'Gerente' , 'Desarrollador' , 'Gerente' , 'Analista', None],
    'FechaIngreso': ['2021-01-15' , '2020/03/12' , '2022-07-01' , '2021/12/01' , None, '2020-03-12']
}

df = pd.DataFrame(data)
print("Información completa:")
print(df)

# 1.	Identificar y contar los valores faltantes en cada columna.
valores_faltantes = df.isnull().sum()
print("\nValores faltantes por columna:")
print(valores_faltantes)

# Reemplazar 'NaN' en 'Edad' por la media de edades
df['Edad'] = df['Edad'].fillna(df['Edad'].mean())
print("\nTabla con valores de edad inválidos reemplazados por la media:")
print(df)

# 2.	Verificar el formato de la columna 'Salario' para asegurar que todos los valores sean numéricos.
salarios_incorrectos = df[~df['Salario'].str.replace('[^0-9]', '', regex=True).fillna('').str.isdigit()]

# Reemplazar el salario_incorrecto por la media
index = salarios_incorrectos.index
df.loc[index, 'Salario'] = None
df['Salario'] = df['Salario'].astype(np.float64) # Puede usarse también .astype(int)
df['Salario'] = df['Salario'].fillna(df['Salario'].mean())
print('\nTabla con salarios numéricos:')
print(df)

# 3.	Validar y estandarizar el formato de la columna 'Fecha Ingreso' a YYYY-MM-DD.
df['FechaIngreso'] = df['FechaIngreso'].str.replace('/', '-')
df['FechaIngreso'] = pd.to_datetime(df['FechaIngreso'], errors='coerce', format='%Y-%m-%d')
print("\nDatos después de estandarizar 'FechaIngreso':")
print(df)

# 4.	Verificar que no haya valores faltantes en la columna "Cargo". Reemplazar los valores faltantes por equivalente en 'Salario'
# Reemplazar los valores faltantes en 'Cargo' buscando registros con el mismo Salario
for i, row in df[df['Cargo'].isnull()].iterrows():
    # Buscar un registro con el mismo Salario y un Cargo no nulo
    salario = row['Salario']
    posible_cargo = df[(df['Salario'] == salario) & (df['Cargo'].notnull())]['Cargo']

    # Si se encuentra un posible cargo, reemplazar el valor faltante
    if not posible_cargo.empty:
        df.at[i, 'Cargo'] = posible_cargo.iloc[0]

print("\nTabla después de reemplazar valores faltantes en 'Cargo':")
print(df)
```
