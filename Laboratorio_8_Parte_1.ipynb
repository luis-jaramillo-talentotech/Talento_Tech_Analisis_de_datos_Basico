```python
import matplotlib.pyplot as plt

# Datos de ventas
data = {
    'Producto': ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E', 'Producto F', 'Producto G', 'Producto H'],
    'Bogot': [150, 100, 90, 120, 130, 110, 160, 180],
    'Cali': [200, 180, 150, 170, 160, 190, 210, 220],
    'Bucaramanga': [250, 220, 300, 280, 260, 240, 300, 290]
}

productos = data['Producto']
bogota = data['Bogot']
cali = data['Cali']
bucaramanga = data['Bucaramanga']

# Creacin del grfico de lneas
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(productos, bogota, marker='*', label='Bogot', linestyle="-.", linewidth=2)
ax.plot(productos, cali, marker='o', label='Cali', linestyle=":", linewidth=3)
ax.plot(productos, bucaramanga, marker='x', label='Bucaramanga', linestyle="-")

# Configuraciones adicionales
ax.set_xlabel('Productos')
ax.set_ylabel('Ventas')
ax.set_title('Ventas de Productos por Ciudad')
ax.legend()

# Mostrar grfico
plt.tight_layout()
plt.show()
```

-	Realizar un grfico personalizado (aplicar un estilo atractivo) para visualizar la evolucin de las ventas de tres productos (Papa, Cebolla, Tomate) a lo largo de 8 meses en tres ciudades diferentes.

```python
import matplotlib.pyplot as plt

# Datos de ventas de ejemplo para los productos a lo largo de 8 meses en tres ciudades
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto']
productos = ['Papa', 'Cebolla', 'Tomate']

# Ventas ficticias para cada producto en cada ciudad
ventas = {
    'Papa': {'Bogot': [150, 160, 170, 180, 175, 165, 190, 200],
             'Cali': [100, 150, 200, 250, 300, 400, 300, 150],
             'Cartagena': [130, 125, 135, 140, 145, 150, 155, 160]},
    'Cebolla': {'Bogot': [90, 120, 160, 200, 240, 100, 200, 150],
                'Cali': [90, 95, 100, 105, 110, 115, 120, 125],
                'Cartagena': [85, 80, 85, 90, 95, 100, 105, 110]},
    'Tomate': {'Bogot': [200, 210, 215, 220, 230, 240, 235, 250],
               'Cali': [180, 175, 185, 190, 200, 205, 210, 215],
               'Cartagena': [80, 110, 150, 190, 230, 330, 350, 240]}
}

# Aplicacin de un estilo visual atractivo
plt.style.use('Solarize_Light2')
fig, axs = plt.subplots(3, 1, figsize=(9, 9), sharex=True)

# Colores para cada ciudad
colores = {'Bogot': 'tab:blue', 'Cali': 'tab:orange', 'Cartagena': 'tab:green'}

# Generar grficos para cada producto
for i, producto in enumerate(productos):
    for ciudad, color in colores.items():
        axs[i].plot(meses, ventas[producto][ciudad], marker='o', linestyle='-', color=color, label=ciudad)

    axs[i].set_title(f'Evolucin de ventas de {producto} en diferentes ciudades')
    axs[i].set_ylabel('Ventas')
    axs[i].legend()
    axs[i].grid(True, which='both', linestyle='--', linewidth=0.5)

# Configuracin general del grfico
axs[-1].set_xlabel('Meses')
plt.suptitle('Evolucin de Ventas de Productos a lo Largo de 8 Meses', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
