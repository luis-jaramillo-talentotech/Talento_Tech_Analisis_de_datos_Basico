```python
import matplotlib.pyplot as plt

# Datos de ventas
data = {
    'Producto': ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E', 'Producto F', 'Producto G', 'Producto H'],
    'Bogot': [150, 100, 90, 120, 130, 110, 160, 180],
    'Cali': [200, 180, 150, 170, 160, 190, 210, 220],
    'Bucaramanga': [250, 220, 300, 280, 260, 240, 300, 290]
}

productos = data['Producto']
bogota = data['Bogot']
cali = data['Cali']
bucaramanga = data['Bucaramanga']

# Creacin del grfico de lneas
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(productos, bogota, marker='*', label='Bogot', linestyle="-.", linewidth=2)
ax.plot(productos, cali, marker='o', label='Cali', linestyle=":", linewidth=3)
ax.plot(productos, bucaramanga, marker='x', label='Bucaramanga', linestyle="-")

# Configuraciones adicionales
ax.set_xlabel('Productos')
ax.set_ylabel('Ventas')
ax.set_title('Ventas de Productos por Ciudad')
ax.legend()

# Mostrar grfico
plt.tight_layout()
plt.show()
```
